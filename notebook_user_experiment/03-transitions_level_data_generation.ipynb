Putting together all information that can be useful for the flow analysis.

Input:

* transitions.csv (flow_score given by each participant for each playlist that he or she analyzed)

Information from creating playlists:

* info_playlist_low_low_p90_2019-06-14.csv
* info_playlist_low_high_p90_2019-06-14.csv
* info_playlist_high_low_p90_2019-06-14.csv
* info_playlist_low_low_p20_50_2019-06-14.csv
* info_playlist_low_high_p20_50_2019-06-14.csv
* info_playlist_high_low_p20_50_2019-06-14.csv

Output:

* playlists_diff_info_2019-07-04
* transitions_info_complete_2019-07-04




```python
import pandas as pd
import time
TodaysDate = time.strftime("%Y-%m-%d")
```

```python
folder_data_generated = "./data/DATA_USER_EXPERIMENT/data_generated/"
folder_data_processed = "./data/DATA_USER_EXPERIMENT/data_processed/"
```

```python
df_transitions = pd.read_csv(folder_data_generated+"transitions.csv")
```

```python
df_transitions.head()
```

```python
# remove rows with transitions 1,2 and 3 (category 'mood')
df_transitions = df_transitions[~df_transitions['playlist'].isin(['1','2','3'])]
```

```python
# renaming playlist in a more meaningful way

df_transitions['playlist'].replace({'4B': 'low_mood_low_tempo_not_extr','5B': 'low_mood_high_tempo_not_extr',
                                    '6B': 'high_mood_low_tempo_not_extr'},inplace=True)
df_transitions['playlist'].replace({'4': 'low_mood_low_tempo_extr','5': 'low_mood_high_tempo_extr',
                                    '6': 'high_mood_low_tempo_extr'},inplace=True)
```

```python
df_transitions.reset_index(drop = True, inplace=True)
df_transitions.head(10)
```

```python
# opening info about diff

df_low_low_p90 = pd.read_csv(folder_data_generated+"info_playlist_low_low_p90_2019-06-14.csv")
df_low_high_p90 = pd.read_csv(folder_data_generated+"info_playlist_low_high_p90_2019-06-14.csv")
df_high_low_p90 = pd.read_csv(folder_data_generated+"info_playlist_high_low_p90_2019-06-14.csv")

df_low_low_p20_50 = pd.read_csv(folder_data_generated+"info_playlist_low_low_p20_50_2019-06-14.csv")
df_low_high_p20_50 = pd.read_csv(folder_data_generated+"info_playlist_low_high_p20_50_2019-06-14.csv")
df_high_low_p20_50 = pd.read_csv(folder_data_generated+"info_playlist_high_low_p20_50_2019-06-14.csv")
```

```python
df_low_low_p90.head()
```

```python
df_low_high_p20_50.head()
```

```python
# concatenate all

df_playlists_diff_concatenated = pd.concat([df_low_low_p90,df_low_high_p90,df_high_low_p90,df_low_low_p20_50,df_low_high_p20_50,df_high_low_p20_50])
```

```python
df_playlists_diff_concatenated.reset_index(drop = True, inplace=True)
```

```python
df_playlists_diff_concatenated.info()
```

```python
df_playlists_diff_concatenated.playlist.unique()
```

```python
# rename playlists

df_playlists_diff_concatenated['playlist'].replace({'low_low_p90': 'low_mood_low_tempo_extr',
                                                    'low_high_p90': 'low_mood_high_tempo_extr',
                                                    'high_low_p90': 'high_mood_low_tempo_extr', 
                                                    'low_low_p20_50': 'low_mood_low_tempo_not_extr',
                                                    'low_high_p20_50': 'low_mood_high_tempo_not_extr',
                                                    'high_low_p20_50': 'high_mood_low_tempo_not_extr'},inplace=True)
```

```python
df_playlists_diff_concatenated.head()
```

```python
# save info playlists diff 
df_playlists_diff_concatenated.to_csv(folder_data_processed + "playlists_diff_info_"+TodaysDate+".csv", index=False)
```

```python
df_test = pd.read_csv(folder_data_processed + "playlists_diff_info_"+TodaysDate+".csv")
df_test.head()
```

```python
# merge transitions with df_playlists_diff_concatenated

df_transitions_complete = pd.merge(df_transitions, df_playlists_diff_concatenated, on = ['playlist','transitions'])
```

```python

df_transitions_complete.info()
```

```python
df_transitions_complete.sort_values(by=['participant_id','playlist'],inplace=True)
```

```python
df_transitions_complete.to_csv(folder_data_processed + "transitions_info_complete_"+TodaysDate+".csv", index=False)
```
