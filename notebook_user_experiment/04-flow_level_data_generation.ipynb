# Creating new columns and creating categorical features

## From the name of playlists create columns that shows the variation (low, high) and level of high (extreme, not_extreme) for energy, valence and tempo. Create also one column with mood since it can represent both energy and valence since the vary always in the same way.

```python
def new_columns_level_extreme(row):
    """ function to create column with 'not_extreme' when high uses boundaries between 20% and 50% percentaile and
    'extreme otherwise """
    
    row.split('_')[-2]
    if 'not'in row:
        return 'not_extreme'
    else:
        return 'extreme'
```

```python
list_mood = [playlist.split('_')[0] for playlist in df_flow.playlist]
list_tempo = [playlist.split('_')[2] for playlist in df_flow.playlist]

df_flow['var_energy'] = list_mood
df_flow['var_valence'] = list_mood
df_flow['var_mood'] = list_mood
df_flow['var_tempo'] = list_tempo

df_flow['level_extreme']=df_flow['playlist'].apply(lambda row: new_columns_level_extreme(row))
```

```python
df_flow[['playlist','var_energy', 'var_valence','var_mood','var_tempo','level_extreme']].head()
```

## Creating some categorical data at participant level

```python
df_describe = df_flow[['age','gender','hours_spotify','SUM_msi', 'extraversion', 'agreeableness',
       'conscientiousness', 'neuroticism', 'openness', 'companionship',
       'investment']].describe()

df_describe
```

```python
# gender is recoded in male and female

df_flow['gender_coded'] = df_flow['gender'].apply(lambda x: 'male' if x==0 else 'female')
```

```python
# age is divided in two groups: under the median : 'low', above : 'high'

df_flow['age_coded'] = df_flow['age'].apply(lambda x: 'low' if x<=df_describe['age']['50%'] else 'high')
```

```python
def coding_variables_25_75(row, column_name):
    if row < df_describe[column_name]['25%']:
        return 'low'
    elif row > df_describe[column_name]['75%']:
        return'high'
    else: return 'medium'


def coding_variables_50_75(row, column_name):
    if row < df_describe[column_name]['50%']:
        return 'low'
    elif row > df_describe[column_name]['75%']:
        return'high'
    else: return 'medium'
```

```python
# variables coded as 'low' under the 1st quartile, 'medium' between the 1st and the 3rd, and 'high' above the 3rd quartile

list_variable_25_75 = ['SUM_msi', 'extraversion', 'agreeableness',
       'conscientiousness', 'neuroticism', 'openness', 'companionship',
       'investment']

for variable in list_variable_25_75:
    df_flow[variable+'_coded']=df_flow[variable].apply(lambda row: coding_variables_25_75(row,variable))

# rename 'hours_spotify_coded' 

df_flow.rename(columns = {'hours_spotify_coded':'hours_spotify_coded_7_scale'},inplace = True)

# code hours spotify using median and 3rd quartile as boundaries

df_flow['hours_spotify_coded']=df_flow['hours_spotify'].apply(lambda row: coding_variables_50_75(row,'hours_spotify'))
```
